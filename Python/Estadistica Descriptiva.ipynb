```python
iris2 = iris
```

```python
iris2 = iris2.rename(columns={"sepal_length": "sepal_length_dataset2", "sepal_width": "sepal_width_dataset2", "petal_length": "petal_length_dataset2", "petal_width": "petal_width_dataset2", "species": "species_dataset2", "Suma de columnas": "Suma de columnas_dataset2", "petal mod": "petal mod_dataset2"})
```

```python
iris_merge = pd.merge(iris, iris2, how="right", left_on=["sepal_length"], right_on=["sepal_length"],
                     suffixes=("_iris", "_iris2")
                     )
```
